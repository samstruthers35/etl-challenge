```python
from bs4 import BeautifulSoup
import requests
import pymongo
import pandas as pd
```

```python
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)
```

```python
db = client.soccer_db
collection = db.players
```

```python
url = "https://fbref.com/en/country/players/BRA/Brazil-Football-Players"
response = requests.get(url)
soup = BeautifulSoup(response.text, 'lxml')
print(soup.prettify())
```

```python
results = soup.find_all('div', class_='section_content')
results
```

```python
player_names = []
for result in results:
    player_names = player_names + result.find_all('p')
```

```python
print(player_names[100])
```

```python
player_url = player_names[100].find('a')['href'].split('/')[3]
print(player_url)
```

```python
player_info = player_names[100].text.split(' ')
print(player_info)
```

```python
position = player_names[100].text.split('\xa0')[1]
print(position)
```

```python
player_id = []
name_of_player = []
years_played = []
player_position = []
for name in player_names:
    try:
        players = name.text.split(' 2')[0]
        years = name.text.split('\xa0')[0][-9:]
        position = name.text.split('\xa0')[1][-2:]
        ids = name.find('a')['href'].split('/')[3]
        player_id.append(ids)
        name_of_player.append(players)
        years_played.append(years)
        player_position.append(position)
        if (players and years and ids and position):
            post = {
                'name': players,
                'years active': years,
                'player id': ids,
                'position': position
            }

            collection.insert_one(post)
    except IndexError:
        oops = "Oops"
        print(oops)
        player_position.append(oops)
```

```python
print(player_position)
```

```python
print(name_of_player)
```

```python
zipped_list = list(zip(name_of_player, player_id, years_played, player_position))
df = pd.DataFrame(zipped_list)
df = df.rename(columns={0: "Player Name",1:"Players ID",2:"Years Active",3:"Position Played"})
df
```
